## Step 11: Label Encoding the non-numeric columns

```python
from sklearn.preprocessing import LabelEncoder
le  = LabelEncoder()
```

```python
dfle = df
dfle['Species'] = le.fit_transform(dfle['Species'])
```

```python
dfle.head()
```

```python
dfle
```

## Step 12: Separating out the dependent and independent columns

```python
x = dfle.drop('Species',axis='columns')
y = dfle.Species
```

```python
x
```

```python
x.drop('Id',axis='columns',inplace=True)
```

```python
x
```

## Step 13: Scaling the data

```python
from sklearn import preprocessing
min_max = preprocessing.MinMaxScaler()
scaled_df = min_max.fit_transform(x.values)
```

```python
scaled_df
```

```python
cols=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
```

```python
final_df = pd.DataFrame(scaled_df,columns=["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"])
```

## Step 14: Splitting the dataset into test and train set

```python
# split X and y into training and testing sets
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(final_df[cols],
                                                  y,
                                                  test_size = 0.2,
                                                  random_state = 123 )
```

## Step 15: Training the model

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier( max_depth = 4, max_features=2)
model.fit( train_X, train_y )
```

```python
model.fit(train_X, train_y)
```

```python
print("The Accuracy using Decision tree is :", (model.score(test_X, test_y))*100)
```

## Step 17: Cross val score function

```python
from sklearn.model_selection import cross_val_score
```

### Decision tree performance using cross_val_score

```python
cross_val_score(DecisionTreeClassifier(max_depth = 4, max_features=2),train_X, train_y,cv=3)
```

## Step 18: Making predictions on test dataset

```python
y_predicted = model.predict(test_X)
y_predicted
```

## Step 19: Plotting the predicted results

```python
y_pred = pd.DataFrame(y_predicted)
y_pred.value_counts().plot.bar('Flowers')
```

## Step 20: Visualizing the score

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(test_y,y_predicted)
cm
```

```python
import seaborn as sb
plt.figure(figsize=(6,4))
sb.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
import sklearn.metrics as metrics
```

```python
print('Confusion matrix \n')
print(metrics.confusion_matrix(test_y,y_predicted))
print('*'*80)
print('\n')
print('Classification report \n')
print(metrics.classification_report(test_y,y_predicted))
```

## Step 21: Plotting the Decision Tree

```python
from sklearn import tree
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa', 'versicolor', 'virginica']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (15,10), dpi=300)
tree.plot_tree(model,
               feature_names = fn, 
               class_names=cn,
               filled = True);
fig.savefig('imagename.png')
```
